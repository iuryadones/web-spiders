```python
import json
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arquivo = open("dados_2017.json")
leitura = arquivo.read()
dados = json.loads(leitura)
```

```python
frame_data = pd.DataFrame(dados)
frame_data
```

```python
acoes = {}
for i, municipio in enumerate(frame_data.municipio):
    acoes[municipio] = pd.DataFrame(frame_data.acoes[i])
```
